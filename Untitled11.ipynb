```python
import matplotlib.pyplot as plt

# 1. Carregar o arquivo
df_vendas = pd.read_csv('vendas.csv')

# 2. Limpeza: Verificação de Nulos e Duplicados
print("Valores Ausentes:\n", df_vendas.isnull().sum())
print("\nLinhas Duplicadas:", df_vendas.duplicated().sum())

# Estratégia de Limpeza
df_vendas = df_vendas.dropna(subset=['valor_venda', 'quantidade'])
df_vendas = df_vendas.drop_duplicates()

# 3. Cálculo do Faturamento
df_vendas['faturamento'] = df_vendas['valor_venda'] * df_vendas['quantidade']

# 4. Ajuste de Data e Agrupamento por Mês
df_vendas['data_venda'] = pd.to_datetime(df_vendas['data_venda'])
faturamento_mensal = df_vendas.resample('ME', on='data_venda')['faturamento'].sum()

# 5. Agrupamento por Categoria
faturamento_categoria = df_vendas.groupby('categoria')['faturamento'].sum().sort_values(ascending=False)

# --- VISUALIZAÇÕES ---

# Gráfico de Linha: Tendência Mensal
plt.figure(figsize=(10, 5))
faturamento_mensal.plot(kind='line', marker='o', color='royalblue')
plt.title('Faturamento Total por Mês (Q1)')
plt.xlabel('Mês')
plt.ylabel('Faturamento (R$)')
plt.grid(True)
plt.show()

# Gráfico de Barras: Por Categoria
plt.figure(figsize=(10, 5))
faturamento_categoria.plot(kind='bar', color='seagreen')
plt.title('Faturamento por Categoria')
plt.xlabel('Categoria')
plt.ylabel('Faturamento (R$)')
plt.xticks(rotation=45)
plt.show()
```
